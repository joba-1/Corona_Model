```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
infectivity_df = pd.read_csv('datafiles/state_transition_probs/infectivity.csv')
recovery_df = pd.read_csv('datafiles/state_transition_probs/recovery.csv')
hospitalisation_df = pd.read_csv('datafiles/state_transition_probs/hospitalisation.csv')
icu_death_risk_df = pd.read_csv('datafiles/state_transition_probs/death_risk.csv')
to_icu_df = pd.read_csv('datafiles/state_transition_probs/to_icu.csv')
icu_to_hospital_df = pd.read_csv('datafiles/state_transition_probs/icu_to_hospital.csv')
diagnosis_df = pd.read_csv('datafiles/state_transition_probs/diagnosis.csv')
```

```python
diagnosis_df.columns
```

```python
f, s = plt.subplots()
s.plot(infectivity_df['Time(h)'], infectivity_df['Probability_to_infect'])
s.set_ylabel('P(infect)')
s.set_xlabel('Time(h)')
s.set_title('Infectivity')
```

```python
f, s = plt.subplots()
s.plot(recovery_df['time/hours'], recovery_df['recover probability'])
s.set_ylabel('P(recover)')
s.set_xlabel('Time(h)')
s.set_title('Recovery probability')
```

```python
f, s = plt.subplots()
ys = hospitalisation_df.columns[2:]
x = hospitalisation_df.columns[1]
for y in ys:
    s.plot(hospitalisation_df[x], hospitalisation_df[y])
s.set_ylabel('P(hospitalise)')
s.set_xlabel('Time(h)')
s.set_title('Hospitalisation probability')
```

```python
f, s = plt.subplots()
ys = icu_death_risk_df.columns[2:]
x = icu_death_risk_df.columns[1]
for y in ys:
    s.plot(icu_death_risk_df[x], icu_death_risk_df[y])
s.set_ylabel('P(die in ICU)')
s.set_xlabel('Time(h)')
s.set_title('Death in ICU probability')
```

```python
f, s = plt.subplots()
ys = to_icu_df.columns[2:]
x = to_icu_df.columns[1]
for y in ys:
    s.plot(to_icu_df[x], to_icu_df[y])
s.set_ylabel('P(hospital to ICU)')
s.set_xlabel('Time(h)')
s.set_title('Hospital to ICU probability')
```

```python
f, s = plt.subplots()
ys = icu_to_hospital_df.columns[2:]
x = icu_to_hospital_df.columns[1]
for y in ys:
    s.plot(icu_to_hospital_df[x], icu_to_hospital_df[y])
s.set_ylabel('P(ICU to hospital)')
s.set_xlabel('Time(h)')
s.set_title('ICU to hospital probability')
```

```python
f, s = plt.subplots()
s.plot(diagnosis_df['Hours'], diagnosis_df['probability'])
s.set_ylabel('P(diagnosis)')
s.set_xlabel('Time(h)')
s.set_title('Diagnosis probability')
```
