```python
import os 
os.chdir('../') # important if this notebook is not in main but in notebooks folder

import pandas as pd
import argparse
import sys
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
## function that takes as input the different duration outputs and creates violinplots from these

transitions = ['infection_to_recovery', 'infection_to_death', 'infection_to_hospital', \
               'hospital_to_recovery', 'hospital_to_death', 'hospital_to_icu']
transitionNames = ['infection to recovery', 'infection to death', 'infection to hospital', \
               'hospital to recovery', 'hospital to death', 'hospital to icu']

## define figure
fig, axes = plt.subplots(int(len(transitions)/2),2,figsize=(12,12))

## define subfigures
for n,t in enumerate(transitions):
    row = n%2
    col = int(n/2)
    ax= axes[col][row]
    
    ## please specify the input file path, the current one assumes that the durations have been plotted to outputs
    inputfile = "./outputs/no_mitigation_IF_030_" + t + ".csv"
    #inputfile = "./../../checkStuff/violinplots/no_mitigation_IF_030_" + t + ".csv"
    frame = pd.read_csv(inputfile, index_col=0)
    #print(frame.head(5))
    sb.violinplot(data=frame, ax=ax)
    ax.set_title(transitionNames[n])
    ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
    #ax.set_xlabel('time, h')
    #ax.set_ylabel('counts')

plt.tight_layout()
fig.text(0.5, -0.01, 'time, h', ha='center', va='center', fontsize=14) ## set x axis
fig.text(-0.01, 0.5, 'count', ha='center', va='center', rotation='vertical', fontsize=14) ## set y axis label
outputfile = "./outputs/violinplots.png"
#outputfile = "./../../checkStuff/violinplots/violinplots.png"
plt.savefig(outputfile)
plt.show()
#plt.close()
```
